```python
# process a sentence using the model
mydoc = nlp("long live the king")

#how computer says I love you in vectors

print(mydoc.vector) #Averaged vector for the entire sentence
```

```python
len(mydoc.vector)
```

```python
print(mydoc[0].vector) #vector for 'long', the first word in the text
```

```python
 #long live the king = vector of "long" + Vector of "live"+ Vector of "the" + vector of "king" /4
 fulldocvec = (mydoc[0].vector+mydoc[1].vector+mydoc[2].vector+ +mydoc[3].vector)/4
```

```python
from scipy.spatial.distance import cosine

# Extract word vectors
vec1 = mydoc.vector
vec2 = fulldocvec

# Compute cosine similarity
cosine_sim = 1 - cosine(vec1, fulldocvec)

print(f"Cosine Similarity': {cosine_sim:.4f}")
```

## Doc2Vec

Doc2Vec is to use vectors to represent document (more than one word is document)

```python
#Assume each sentence in documents corresponds to a separate document.
documents = ["Dog bites man.", "Man bites dog.", "Dog eats meat.", "Man eats food."]


#some quick and dirty preprocessing
processed_docs = [doc.lower().replace(".","") for doc in documents]
processed_docs

print("Document After Pre-Processing:",processed_docs)


```

```python
#Iterate over each document and initiate an nlp instance.
for doc in processed_docs:
    doc_nlp = nlp(doc) #creating a spacy "Doc" object which is a container for accessing linguistic annotations.

    print("-"*30)
    print("Average Vector of '{}'\n".format(doc),doc_nlp.vector)#this gives the average vector of each document
    for token in doc_nlp:
        print()
        print(token.text,token.vector)#this gives the text of each word in the doc and their respective vectors.
```

```python
sentence1 = nlp(processed_docs[0]) #dog bite man
sentence2 = nlp(processed_docs[1]) # man bite dog

# Compute sentence similarity
sentence_similarity = sentence1.similarity(sentence2)

print(f"Sentence Similarity: {sentence_similarity:.4f}")
```
