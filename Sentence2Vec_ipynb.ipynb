```python
from sentence_transformers import SentenceTransformer

# Load InferSent model
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2') #load your model here
```

```python
#Assume each sentence in documents corresponds to a separate document.
documents = ["Dog bites man.", "Man bites dog.", "Dog eats meat.", "Man eats food."]


#some quick and dirty preprocessing
processed_docs = [doc.lower().replace(".","") for doc in documents]
processed_docs

print("Document After Pre-Processing:",processed_docs)


```

```python
# Generate embeddings
embeddings0 = model.encode(processed_docs[0])

print("InferSent Embeddings:\n", embeddings0)
```

```python
# Generate embeddings
embeddings00 = model.encode(documents[0])

print("InferSent Embeddings:\n", embeddings00)
```

```python
from scipy.spatial.distance import cosine

# Extract word vectors
vec1 = embeddings00
vec2 = embeddings0

# Compute cosine similarity
cosine_sim = 1 - cosine(embeddings0, embeddings00)

print(f"Cosine Similarity': {cosine_sim:.4f}")
```
